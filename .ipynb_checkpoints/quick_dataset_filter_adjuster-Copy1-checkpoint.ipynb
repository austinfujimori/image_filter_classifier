```python
import torch
import torchvision
import matplotlib.pyplot as plt

import torchvision.transforms as transforms
from torch.utils.data import random_split
from torch.utils.data import DataLoader
import torch.nn.functional as F

import os
import random
from PIL import Image, ImageEnhance, ImageFilter

from IPython.display import Image as ImagePy
from IPython.core.display import HTML 
from collections import Counter


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
base_dir = 'dataset'

camera_filters = ["warmth"]

camera_filtered_dir = os.path.join(base_dir, 'camera_filtered')
os.makedirs(camera_filtered_dir, exist_ok=True)

def apply_varied_warmth_filter(img):
    enhancer = ImageEnhance.Color(img)
    factor = random.uniform(1.5, 1.9)
    warmth_img = enhancer.enhance(factor)
    return warmth_img


def apply_camera_filter(img, filter_name):
    if filter_name == "warmth":
        return apply_varied_warmth_filter(img)
    else:
        return img

def process_and_save_images(source_dir, target_dir, filter_name, cap):

    all_files = [f for f in os.listdir(source_dir) if f.lower().endswith(('png', 'jpg', 'jpeg'))]
    selected_files = random.sample(all_files, min(cap, len(all_files)))
    
    for file in selected_files:
        source_path = os.path.join(source_dir, file)
        with Image.open(source_path) as img:
            filtered_img = apply_camera_filter(img, filter_name)
            target_file_dir = os.path.join(target_dir, filter_name)
            os.makedirs(target_file_dir, exist_ok=True)
            filtered_img.save(os.path.join(target_file_dir, file))


for subdir, cap in [('train', 86744), ('val', 10954)]:
    source_subdir = os.path.join(base_dir, 'unfiltered', subdir)
    target_subdir = os.path.join(camera_filtered_dir, subdir)
    for filter_name in camera_filters:
        print("Adding " + filter_name + " " + subdir)
        process_and_save_images(source_subdir, target_subdir, filter_name, cap)
```

```python
base_dir = 'dataset'

camera_filters = ["black_and_white"]

camera_filtered_dir = os.path.join(base_dir, 'camera_filtered')
os.makedirs(camera_filtered_dir, exist_ok=True)

from PIL import Image, ImageEnhance, ImageOps
import random

def apply_varied_black_and_white_filter(img):
    # Convert to grayscale
    bw_img = img.convert("L")

    # Optionally, introduce random noise
    pixels = bw_img.load()  # Ensure we're working with the grayscale image
    for y in range(bw_img.size[1]):
        for x in range(bw_img.size[0]):
            if random.random() < 0.05:  # Adjust noise frequency as needed
                # Ensure we're correctly adjusting the grayscale pixel value
                current_pixel = pixels[x, y]
                noise = random.randint(-30, 30)
                new_pixel = max(0, min(255, current_pixel + noise))
                pixels[x, y] = new_pixel  # Update the pixel value

    return bw_img




def apply_camera_filter(img, filter_name):
    if filter_name == "black_and_white":
        return apply_varied_black_and_white_filter(img)
    else:
        return img

def process_and_save_images(source_dir, target_dir, filter_name, cap):

    all_files = [f for f in os.listdir(source_dir) if f.lower().endswith(('png', 'jpg', 'jpeg'))]
    selected_files = random.sample(all_files, min(cap, len(all_files)))
    
    for file in selected_files:
        source_path = os.path.join(source_dir, file)
        with Image.open(source_path) as img:
            filtered_img = apply_camera_filter(img, filter_name)
            target_file_dir = os.path.join(target_dir, filter_name)
            os.makedirs(target_file_dir, exist_ok=True)
            filtered_img.save(os.path.join(target_file_dir, file))


for subdir, cap in [('train', 86744), ('val', 10954)]:
    source_subdir = os.path.join(base_dir, 'unfiltered', subdir)
    target_subdir = os.path.join(camera_filtered_dir, subdir)
    for filter_name in camera_filters:
        print("Adding " + filter_name + " " + subdir)
        process_and_save_images(source_subdir, target_subdir, filter_name, cap)
```

```python
base_dir = 'dataset'

camera_filters = ["cool"]

camera_filtered_dir = os.path.join(base_dir, 'camera_filtered')
os.makedirs(camera_filtered_dir, exist_ok=True)

def apply_varied_cool_filter(img):
    enhancer = ImageEnhance.Color(img)
    factor = random.uniform(0.1, 0.7)
    cool_img = enhancer.enhance(factor)

    # Create a blue overlay
    blue_overlay = Image.new('RGB', cool_img.size, (0, 145, 255))

    alpha = random.uniform(0.01, 0.1)
    cool_img_with_blue_tint = Image.blend(cool_img, blue_overlay, alpha=alpha)

    return cool_img_with_blue_tint


def apply_camera_filter(img, filter_name):
    if filter_name == "cool":
        return apply_varied_cool_filter(img)
    else:
        return img

def process_and_save_images(source_dir, target_dir, filter_name, cap):

    all_files = [f for f in os.listdir(source_dir) if f.lower().endswith(('png', 'jpg', 'jpeg'))]
    selected_files = random.sample(all_files, min(cap, len(all_files)))
    
    for file in selected_files:
        source_path = os.path.join(source_dir, file)
        with Image.open(source_path) as img:
            filtered_img = apply_camera_filter(img, filter_name)
            target_file_dir = os.path.join(target_dir, filter_name)
            os.makedirs(target_file_dir, exist_ok=True)
            filtered_img.save(os.path.join(target_file_dir, file))


for subdir, cap in [('train', 86744), ('val', 10954)]:
    source_subdir = os.path.join(base_dir, 'unfiltered', subdir)
    target_subdir = os.path.join(camera_filtered_dir, subdir)
    for filter_name in camera_filters:
        print("Adding " + filter_name + " " + subdir)
        process_and_save_images(source_subdir, target_subdir, filter_name, cap)
```

```python


def display_random_images(folder_path, num_images=20, images_per_row=5):
    all_files = os.listdir(folder_path)
    image_files = [f for f in all_files if f.endswith(('.jpg', '.jpeg', '.png'))]
    selected_files = random.sample(image_files, min(num_images, len(image_files)))
    num_rows = (len(selected_files) + images_per_row - 1) // images_per_row

    plt.figure(figsize=(20, 4 * num_rows))
    for idx, file_name in enumerate(selected_files):
        img_path = os.path.join(folder_path, file_name)
        image = Image.open(img_path)
        plt.subplot(num_rows, images_per_row, idx + 1)
        plt.imshow(image)
        plt.title(file_name)
        plt.axis('off')
    plt.tight_layout()
    plt.show()


folder_path = 'dataset/camera_filtered/train/cool'
display_random_images(folder_path)
```
